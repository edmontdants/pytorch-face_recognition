```
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```
annotation_path = "/content/pytorch-face_recognition/dataset/CelebA/list_bbox_celeba.txt"
train_path = "/content/pytorch-face_recognition/dataset/CelebA/train.txt"
valid_path = "/content/pytorch-face_recognition/dataset/CelebA/valid.txt"
test_path = "/content/pytorch-face_recognition/dataset/CelebA/test.txt"
```

```
n_train = 10000
n_valid = 1000
n_test = 5000
```

```
f = open(annotation_path, 'r')
f_train = open(train_path, 'w')
f_valid = open(valid_path, 'w')
f_test = open(test_path, 'w')

line = f.readline() # Total data

f_train.write(str(n_train)+"\n")
f_valid.write(str(n_valid)+"\n")
f_test.write(str(n_test)+"\n")

line = f.readline()
f_train.write(line)
f_valid.write(line)
f_test.write(line)
```

```
for n in range(n_train):
    line = f.readline()
    f_train.write(line)

f_train.close()
```

```
for n in range(n_valid):
    line = f.readline()
    f_valid.write(line)

f_valid.close()
```

```
for n in range(10000 - n_valid):
    line = f.readline()
```

```
for n in range(n_test):
    line = f.readline()
    f_test.write(line)

f.close()
f_test.close()
```
